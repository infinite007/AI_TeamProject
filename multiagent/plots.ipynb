```python
#Code based on matplotilib documentation gallery: https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

labels = ['Small layout\nRandom ghosts','Medium layout\nRandom ghosts','Small layout\nAdversarial ghosts','Medium layout\nAdversarial ghosts']
agents = ["Alpha-beta","Expectimax","MC Tree Search"]
ab_means = [1.0,0.9,0.45,0.45]
exp_means = [0.95,1.0,0.3,0.3]
mc_means = [0.95,1.0,0.8,0.9]

x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, ab_means, width,label=agents[0], color='dodgerblue')
rects2 = ax.bar(x, exp_means, width,label=agents[1], color='coral')
rects3 = ax.bar(x + width, mc_means,width, label=agents[2], color='mediumseagreen')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Win percentage')
ax.set_title('Comparison of Pac-Man agents')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.tick_params(axis='both', which='major', labelsize=8)
leg = ax.legend(bbox_to_anchor=(0.8, 0.9),fontsize='small')
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',fontsize=8)


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
plt.savefig("figures/comparison.png")
```
